```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')        # Head-less 설정, 브라우저를 열지 않음
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('C:\\PROJECT\\crawler\\driver\\chromedriver', options=options)
driver.get(url="https://developer.android.com/reference/android/R.animator")
```

```python
URL = []

for i in range(4,247):
    url_tmp = driver.find_elements(By.XPATH,f'/html/body/section/devsite-book-nav/nav/div[2]/div[2]/ul/li[4]/div/ul/li[{i}]/div/ul/*')
    cnt_tmp = len(url_tmp)
    for j in range(2,cnt_tmp+1):
        url_tmp_1 = driver.find_elements(By.XPATH,f'/html/body/section/devsite-book-nav/nav/div[2]/div[2]/ul/li[4]/div/ul/li[{i}]/div/ul/li[{j}]/div/ul/*')
        cnt_tmp_1 = len(url_tmp_1)
        for k in range(1,cnt_tmp_1):
            url_tmp_2 = driver.find_element(By.XPATH,f'/html/body/section/devsite-book-nav/nav/div[2]/div[2]/ul/li[4]/div/ul/li[{i}]/div/ul/li[{j}]/div/ul/li[{k}]/a')
            URL.append(url_tmp_2.get_attribute('href'))
```

```python
API = []

for i in URL[3:]:
    header = i.split('/')[-1]
    driver.get(url=i)
    try:
        method = len(driver.find_elements(By.XPATH, '//*[@id="constants"]/tbody/*'))
        for i in range(2,method+1):
            api = driver.find_element(By.XPATH,f'//*[@id="constants"]/tbody/tr[{i}]/td[2]/code').text
            str_tmp = header+"."+api
            if str_tmp not in API:
                API.append(str_tmp)
    except:
        continue
```

```python
with open("./api.txt","w") as file:
    file.write(str(API))
```
